### Part of speech tagging

```python
def pos_tag(sent):
    doc = nlp(sent)
    return [(token.text, token.lemma_, token.tag_) for token in doc]
```

```python
test = df_final.iloc[1,13]
print(test)
Counter([(lemma,tag)for txt,lemma,tag in pos_tag(test) if tag.startswith("VB")]).most_common(2)
```

## Overall verb frequency

```python
verb_counter = Counter()
for i,row in df_final.iterrows():
    verbs = Counter([lemma for _,lemma,tag in pos_tag(row[13]) if tag.startswith("VB")])
    for key,value in verbs.items():
        verb_counter[key] += value
    if i % 2000==0:
        print(i)
```

```python
verb_counter.most_common(50)
```
